```python
# 첫행을 날리자 (skiprows)
df1 = pd.read_excel('Data05.xlsx',skiprows=1) # 맨위1줄을 스킵
```

```python
df1.shape   # (93, 27) : 93개 행, 27개 열
```

```python
df1.columns # '구분'부터 '제품명' 열 까지는 정상
```

```python
# df1.iloc[0] # 0번째 row 가져옴
# df1.iloc[0:5] # 0번째부터 5번째까지 row 가져옴
df1.iloc[:] # 전체 row 가져옴
```

```python
# df1.iloc[:, 0]  # 전체 row 에서 0번째 column을 가져옴 : iloc[index, column]
# df1.iloc[:, 8:]   # 전체 row 에서 8번째 부터 마지막 column을 가져옴
df1.iloc[:, 8:-2]   # 전체 row 에서 8번째 부터 마지막 2개 앞까지 column을 가져옴
```

```python
df1.iloc[:, 8:-2].T # 행과열 전환(Transfer)
```

```python
df2 = df1.iloc[:, 8:-2]
df2.stack()
pd.DataFrame(df2.stack()).reset_index
```
