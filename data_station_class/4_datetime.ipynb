```python
import pandas as pd
```

```python
df2 = pd.read_csv('Data02.csv')
print(df2.shape)
df2.head()
```

```python
df2['배송시작일']   # dtype: object 문자로 인식
```

```python
pd.to_datetime(df2['배송시작일'])   # TypeError : to_datetime으로 처리할수 있는 데이터 형태가 있다.
```

```python
df2['배송시작일']   # 앞에 있는 2019-01-07 만 빼오자
```

```python
def date_function(x):
    return x[0:10]

df2['배송시작일'].apply(date_function)  # df2['배송시작일'] 의 데이터 시리즈가 한줄한줄 date_function함수의 인자에 들어감
```

```python
pd.to_datetime(df2['배송시작일'].apply(date_function))  # dtype: datetime64[ns]
df2['배송시작일(datetime)'] = pd.to_datetime(df2['배송시작일'].apply(date_function))
```

```python
df2['배송완료일(datetime)'] = pd.to_datetime(df2['배송완료일'].apply(date_function))
df2
```

```python
df2['배송완료일(datetime)']-df2['배송시작일(datetime)'] # timedelta64[ns] : 시간의 간격을 의미하는 데이터 타입 형태
df2['배송시간'] = df2['배송완료일(datetime)']-df2['배송시작일(datetime)']
df2.head()
```

```python
df2['배송시간'].describe()  # 평균, 최대, 최소 값 개수 등 확인
```

```python
df2['주문일']   # dtype : int64
```

```python
pd.to_datetime(df2['주문일'])   # 1970년 1월 1일부터 시작되는 마이크로 초 형태로 인식
```

```python
df2['주문일']   # 중간에 '-' 삽입

def date_function2(x):
    # return x[0:4]   # 문자형이 아닌 정수형을 슬라이싱으로 가져올수 없음
    result = str(x)
    return  result[0:4] + '-' + result[4:6] + '-' + result[6:8] # 슬라이싱을 이런식으로 붙여 쓸수 있다

df2['주문일'].apply(date_function2) # 문자형
pd.to_datetime(df2['주문일'].apply(date_function2)) # 날짜형
```
