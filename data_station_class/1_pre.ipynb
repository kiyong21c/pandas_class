```python
# 정렬
df1.sort_values(by='입고수량')  # 입고수량 값을 오름차순으로 정렬
df1.sort_values(by='입고수량', ascending=False)  # 입고수량 값을 내림차순으로 정렬
```

```python
df1.sort_values(by=['입고수량','출고수량']) # 입고수량에 대해서 먼저 정렬 후 출고수량에 대해서 정렬
df1.sort_values(by=['입고수량','출고수량'], ascending=[True,False]) # 입고수량에 대해서 오름차순 후 출고수량에 내림차순 정렬
```
